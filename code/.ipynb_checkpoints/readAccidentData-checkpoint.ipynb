```python
import pandas as pd
import glob
import numpy as np
from datetime import datetime

from zipfile import ZipFile as ZF
import os

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from matplotlib.widgets import Slider, Button, RadioButtons
```

```python
data_dir = "../data/_rawdata"
lookup_path = "../data/_rawdata/lookup"
lookup_file = "DEFS2009.xls"
```

```python
accident_DF = pd.DataFrame()
accident_aux_DF = pd.DataFrame()
```

```python
# Read accident data from all the zip files
accident_DF = pd.concat(\
    [pd.read_csv(ZF(os.path.join(data_dir,file)).open("accident.csv")) for file in os.listdir(data_dir) \
     if file.endswith("zip")], \
        ignore_index = True)

```

```python
# Read accident aux data from all the zip files
accident_aux_DF = pd.concat(\
    [pd.read_csv(ZF(os.path.join(data_dir,file)).open("acc_aux.csv")) for file in os.listdir(data_dir) \
     if file.endswith("zip")], \
        ignore_index = True)

```

```python
# write the data into csv for further use
accident_DF.to_csv(os.path.join(data_dir,"accident_allyears.csv"), index = False)

# write the data into csv for further use
accident_aux_DF.to_csv(os.path.join(data_dir,"accident_aux_allyears.csv"), index = False)
```

```python
# Read Lookup data into lookup dataframe for HOUR, MINUTE and DAY_WEEK
hour_DF = pd.read_excel(os.path.join(lookup_path, lookup_file), "HOUR")
# Rename column ID to HOUR and DEF to HOUR_DEF
hour_DF.rename(columns = {"ID" : "HOUR","Def" : "HOUR_DEF" }, inplace = True)

min_DF = pd.read_excel(os.path.join(lookup_path, lookup_file), "MINUTE")
# Rename column ID to HOUR and DEF to HOUR_DEF
min_DF.rename(columns = {"ID" : "MINUTE","DEF" : "MINUTE_DEF" }, inplace = True)

day_week_DF = pd.read_excel(os.path.join(lookup_path, lookup_file), "DAY_WEEK")
# Rename column ID to HOUR and DEF to HOUR_DEF
day_week_DF.rename(columns = {"ID" : "DAY_WEEK","Def" : "DAY_WEEK_DEF" }, inplace = True)
```

```python
print(accident_DF.columns)
print(len(accident_DF))
```

```python
accident_DF.head()
```

```python
accident_DF['ACC_DATE'] = accident_DF['YEAR'].astype(str) + \
                            "-"+accident_DF['MONTH'].astype(str) + \
                            "-"+accident_DF['DAY'].astype(str)
accident_DF["ACC_DATE"] = accident_DF["ACC_DATE"].apply(lambda r: datetime.strptime(r , '%Y-%m-%d'))

accident_DF['ACC_DATE'].head()
```

```python
acc_dt_view = accident_DF[["STATE","ST_CASE","ACC_DATE","YEAR","MONTH","DAY","DAY_WEEK","HOUR","MINUTE",\
                           "FATALS","DRUNK_DR","PERSONS","PEDS","SCH_BUS" ]]
acc_dt_view.head()
```

```python
print(hour_DF.columns)
print(min_DF.columns)
print(day_week_DF.columns)
```

```python
# Add columns to expand the HOUR, MInute, and Day of Week
acc_dt_view = acc_dt_view.merge(hour_DF, on ="HOUR", how = "inner")
acc_dt_view = acc_dt_view.merge(min_DF, on ="MINUTE", how = "inner")
acc_dt_view = acc_dt_view.merge(day_week_DF, on ="DAY_WEEK", how = "inner")
```

```python
acc_dt_view.head()
```

```python
# CLASSIFY DAYS INTO WEEKDAY AND WEEKEND
acc_dt_view.loc[(acc_dt_view['DAY_WEEK'].isin([1,6,7])),"DAY_CLASSIFIER"] = "WEEKEND"
acc_dt_view.loc[(acc_dt_view['DAY_WEEK'].isin([2,3,4,5])),"DAY_CLASSIFIER"] = "WEEKDAY"
```

```python
acc_dt_view.head()
```

```python
# CREATE TIME BINS
time_bins = [0,6,12,16,20,24,100]

time_bin_lbls = ["12AM TO 6AM","6AM TO 12PM","12PM TO 4PM", "4PM TO 8PM","8PM TO 12AM","NOT RECORDED"]
```

```python
acc_dt_view['TIME_GRPS'] = pd.cut(acc_dt_view['HOUR'],time_bins, labels = time_bin_lbls)
```

```python
acc_dt_view.head()
```

```python
# DAY_NIGHT Classifier

acc_dt_view.loc[(acc_dt_view['HOUR'].isin(np.arange(6,19,1))),"DAY_NIGHT_CLASSIFIER"] = "DAY TIME"
acc_dt_view.loc[(~acc_dt_view['HOUR'].isin(np.arange(6,19,1))),"DAY_NIGHT_CLASSIFIER"] = "NIGHT TIME"
acc_dt_view.loc[(acc_dt_view['HOUR'].isin([99])),"DAY_NIGHT_CLASSIFIER"] = "NOT KNOWN"

acc_dt_view.head()
```

```python
acc_dt_Y_byWkCls = acc_dt_view[["ST_CASE","YEAR","FATALS","DAY_CLASSIFIER"]].\
                groupby(['YEAR',"DAY_CLASSIFIER"]).\
                agg({"ST_CASE" : "count",'FATALS':'sum'})


acc_dt_Y_byWkCls.sort_values(by = ['YEAR','FATALS'], ascending = [True,True], inplace = True)
acc_dt_Y_byWkCls.reset_index(inplace = True)

acc_dt_Y_byWkCls.head()
```

```python
acc_dt_Y_byTGrps = acc_dt_view[["ST_CASE","YEAR","FATALS","TIME_GRPS"]].\
                groupby(['YEAR',"TIME_GRPS"]).\
                agg({"ST_CASE" : "count",'FATALS':'sum'})

acc_dt_Y_byTGrps.sort_values(by = ['FATALS'], ascending = [False], inplace = True)
acc_dt_Y_byTGrps.reset_index(inplace = True)

acc_dt_Y_byTGrps.head()
```

```python
acc_dt_Y_byday = acc_dt_view[["ST_CASE","YEAR","FATALS","DAY_WEEK_DEF"]].\
                groupby(['YEAR',"DAY_WEEK_DEF"]).\
                agg({"ST_CASE" : "count",'FATALS':'sum'})


acc_dt_Y_byday.sort_values(by = ['YEAR','FATALS'], ascending = [True,False], inplace = True)
acc_dt_Y_byday.reset_index(inplace = True)

acc_dt_Y_byday.head()
```

```python
acc_dt_Y_byday_time = acc_dt_view[["ST_CASE","YEAR","FATALS","DAY_NIGHT_CLASSIFIER"]].\
                groupby(['YEAR',"DAY_NIGHT_CLASSIFIER"]).\
                agg({"ST_CASE" : "count",'FATALS':'sum'})


acc_dt_Y_byday_time.sort_values(by = ['YEAR','FATALS'], ascending = [True,False], inplace = True)
acc_dt_Y_byday_time.reset_index(inplace = True)

acc_dt_Y_byday_time.head()
```

```python


with sns.color_palette('Spectral_r', 8):
    
    
    #----------------------------------PLOTS expanding on Day classifiers----------------------------------------------
    fig1,axs1 = plt.subplots(1,2, figsize = (11,6), frameon = True)
    g1 = sns.barplot(x = "YEAR", y = "FATALS", data = acc_dt_Y_byWkCls, hue = "DAY_CLASSIFIER", ax= axs1[0])
    g1.legend(loc='center left', bbox_to_anchor=(2.2, 0.75), ncol=1)
    ax1[0].set((xlabel='Year', ylabel='# of Fatalities by Weekend / Weekday')

    g2 = sns.barplot(x = "YEAR", y = "FATALS", data = acc_dt_Y_byday, hue = "DAY_WEEK_DEF", ax = axs1[1])
    g2.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)
    ax1[0].set((xlabel='Year', ylabel='# of Fatalities by days of week')
    

    
    #----------------------------------PLOTS expanding on Time classifiers----------------------------------------------
    fig2,axs2 = plt.subplots(1,2, figsize = (11,6))
    g3 = sns.barplot(x = "YEAR", y = "FATALS", data = acc_dt_Y_byday_time, hue = "DAY_NIGHT_CLASSIFIER", \
                      ax = axs2[0])
    g3.legend(loc='center left', bbox_to_anchor=(2.20, 0.75), ncol=1)
    
    g4 = sns.barplot(x = "YEAR", y = "FATALS", data = acc_dt_Y_byTGrps, hue = "TIME_GRPS", ax = axs2[1])
    g4.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)

    

plt.show()
```

```python
# Create a rolling average of Fatalities by date
acc_GDF_dt = accident_DF[["ACC_DATE","ST_CASE","FATALS"]].\
                        groupby(['ACC_DATE']).\
                        agg({"ST_CASE" : "count",'FATALS':'sum'})

acc_GDF_dt.reset_index(inplace = True)
acc_GDF_dt.head()
```

```python
# Moving Avg for every 15 days
acc_movAvg_DF = acc_GDF_dt[['ACC_DATE','FATALS']].rolling(window = 30, on = 'ACC_DATE').mean()

# the first 14 values will be NAN values, so drop them
acc_movAvg_DF.dropna(how="any", inplace = True)

# change the index start to 0 onwards
acc_movAvg_DF.set_index(np.arange(0,len(acc_movAvg_DF),1), inplace = True)

acc_movAvg_DF.head()
```

### Create a trend plot to visualize fatalities over time

```python
# Plot a moving average of fatalities in 2015 to 2017

x_axis1 = acc_movAvg_DF['ACC_DATE'].apply(lambda d : mpl.dates.date2num(d))
#x_axis1 = temp_movAvg_DF['days']
y_axis1 = acc_movAvg_DF['FATALS'] 

#mask is boolean values that tells plot to use or ignore values
mask = ~np.isnan(y_axis1)

x_axis1 = x_axis1[mask]
y_axis1 = y_axis1[mask]
```

```python
%matplotlib notebook

zoomval = 0
g = sns.lineplot(x_axis1,y_axis1, color = '#0098BF')

ax1 = g.axes
ax1.set_xticklabels(ax1.get_xticks())

labels = [item.get_text() for item in ax1.get_xticklabels()]
labels = [mpl.dates.num2date(float(d)).strftime("%Y-%b") for d in labels]
ax1.set_xticklabels(labels, rotation = 20)

ax1.set(xlabel='Years', ylabel='Avg. Fatalities since 2015', title = "Average Fatalities from 2015 to 2017")

axcolor = 'lightgoldenrodyellow'
axSlider = plt.axes([0.25, 0.01, 0.55, 0.02], facecolor = axcolor)
zoom = Slider(axSlider, 'ZoomView', 0, 1, 0.5, 0.1, dragging = True)

def update(val):
    #amp = samp.val
    zoomval = zoom.val
        
zoom.on_changed(update)

fig = plt.gcf()
fig.set_size_inches(9, 7)




plt.show()
```
