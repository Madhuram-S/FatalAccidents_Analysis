```python
import os
import urllib.request as ur
import requests
from ftplib import FTP
import zipfile
import urllib.parse
#import wget 
import pandas as pd
```

```python

FTP_HOST = "ftp.nhtsa.dot.gov"
OUT_DIR  = "../data/_rawdata"   
Years=[2015,2016,2017]
```

```python

ftp = FTP(FTP_HOST)
ftp.login()
# get list of .zip files

for year in Years:
    FTP_DIR  = "/FARS/"+str(year)+"/National"
    print(FTP_DIR)
    ftp.cwd(FTP_DIR)
    files = ftp.nlst()
    print(files)
    files = [f for f in files if f.lower().endswith("csv.zip")]
    num=len(files)
    for i, fname in enumerate(files, 1):
        print("Downloading {} ({}/{}) ... ".format(fname, i, num), end='')
        
        local_file = os.path.join(OUT_DIR, fname)
        print(local_file)
        
        with open(local_file, "wb") as outf:
            ftp.retrbinary("RETR "+fname, outf.write)
        print("done!")
        
        outf.close()
ftp.close()
```
